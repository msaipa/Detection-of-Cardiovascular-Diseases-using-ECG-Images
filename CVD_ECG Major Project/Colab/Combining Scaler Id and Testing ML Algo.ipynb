# Standardized and PCA-transformed data

Standardization: Standardization, also known as z-score normalization, involves rescaling the features of your data to have a mean of zero and a standard deviation of one. . This process is achieved by subtracting the mean of each feature and then dividing by the standard deviation. This means that after standardization, one unit on the new scale corresponds to one standard deviation in the original data, regardless of the original scale of the data.

Data standardization is an important technique that is mostly performed as a pre-processing step before inputting data into many machine learning models, to standardize the range of features of an input data set.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Assuming 'test_final' is your DataFrame and each row represents a sample, and each column represents a feature

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to your data and transform the data
standardized_data = scaler.fit_transform(test_final.iloc[:, :-1])  # Assuming last column is the target column

# Perform PCA and choose components as 100
pca = PCA(n_components=100)
x_pca = pca.fit_transform(standardized_data)
x_pca = pd.DataFrame(x_pca)

# 'standardized_data' now contains your standardized dataset
print("Standardized and PCA-transformed data:")
print(x_pca)
```

# Computing the covariance matrix

Variance covariance matrix is defined as a square matrix where the diagonal elements represent the variance and the off-diagonal elements represent the covariance. The covariance between two variables can be positive, negative, and zero. A positive covariance indicates that the two variables have a positive relationship whereas negative covariance shows that they have a negative relationship. If two elements do not vary together then they will display a zero covariance

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA

# Assuming test_final is your DataFrame
# Assuming the data is already standardized

# Initialize PCA with desired number of components
pca = PCA(n_components=100)

# Fit PCA to your data
x_pca = pca.fit_transform(result_df.iloc[:, 0:-1])

# Convert the transformed data into a DataFrame
x_pca = pd.DataFrame(x_pca)

# Compute the covariance matrix of the original data
cov_matrix = np.cov(result_df.iloc[:, 0:-1], rowvar=False)

print("Covariance Matrix:")
print(cov_matrix)
print('cov_matrix:', cov_matrix.shape)
```

# **Compute eigenvectors and eigenvalues**

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA

# Assuming test_final is your DataFrame
# Assuming the data is already standardized

# Initialize PCA with desired number of components
pca = PCA(n_components=100)

# Fit PCA to your data
x_pca = pca.fit_transform(test_final.iloc[:, 0:-1])

# Convert the transformed data into a DataFrame
x_pca = pd.DataFrame(x_pca)

# Calculate covariance matrix
cov_matrix = np.cov(x_pca, rowvar=False)

# Compute eigenvectors and eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
print('Eigen values Shape:', eigenvalues.shape)
print('Eigen Vector Shape:', eigenvectors.shape)
```

Step-4

# Sorted eigenvalues and eigenvectors

PC1 is obtained by selecting the first column (index 0) from the sorted_eigenvectors matrix.PC2 is obtained by selecting the second column (index 1) from the sorted_eigenvectors matrix.

```python
import numpy as np

# Compute eigenvectors and eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort eigenvalues and eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]  # Reverse the order to get descending order
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

print("Sorted Eigenvalues:")
print(sorted_eigenvalues)
print("\nSorted Eigenvectors:")
print(sorted_eigenvectors)
# Assuming you've already computed and sorted the eigenvalues and eigenvectors
# sorted_eigenvalues and sorted_eigenvectors

# First Principal Component (PC1)
first_pc = sorted_eigenvectors[:, 0]

# Second Principal Component (PC2)
second_pc = sorted_eigenvectors[:, 1]

print("First Principal Component:")
print(first_pc)

print("\nSecond Principal Component:")
print(second_pc)
# Create a DataFrame to display PC1 and PC2 in tabular form
pc_df = pd.DataFrame({'PC1': first_pc, 'PC2': second_pc})
print(pc_df)
```

Step-6

# Gaussian distribution

We calculate the **mean (mu) and standard deviation (sigma)** of the **first principal component** from the PCA-transformed data x_pca.
We generate data points along the x-axis centered around the mean, using a range that covers several standard deviations from the mean to capture most of the distribution.
We use these mu and sigma values to calculate the probability density function (PDF) for the Gaussian distribution using norm.pdf.
Annotations for the mean and standard deviation are added to the plot using plt.text.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Assuming 'x_pca' contains the PCA-transformed data

# Calculate the mean and standard deviation of the first principal component
mu = np.mean(x_pca.iloc[:, 0])
sigma = np.std(x_pca.iloc[:, 0])

# Generate data points along the x-axis
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 1000)

# Calculate the probability density function (PDF) for the Gaussian distribution
pdf = norm.pdf(x, mu, sigma)

# Plot the Gaussian curve
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, color='blue', linewidth=2, label='Gaussian Curve')

# Add mean and standard deviation annotations
plt.text(mu, norm.pdf(mu, mu, sigma), f'$\mu={mu:.2f}$', verticalalignment='bottom', horizontalalignment='right', fontsize=12)
plt.text(mu + sigma**2, norm.pdf(mu + sigma**2, mu, sigma), f'$\sigma^2={sigma**2:.2f}$', verticalalignment='bottom', horizontalalignment='right', fontsize=12)

# Add labels and title
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Gaussian Distribution')

# Add grid lines
plt.grid(True)

# Add legend
plt.legend()

# Show plot
plt.show()
```

The scatter plot shows the relationships between observations (data points) and the new variables (the principal components).

The code generates a scatter plot to visualize the relationship between the first two principal components obtained from PCA, with each point colored according to its corresponding target label. # **Cumulative Variance Explained**


Cumulative Variance Explained (CVE) is a concept commonly used in statistics and data analysis, particularly in the context of Principal Component Analysis (PCA) or Factor Analysis. It represents the total amount of variation in a dataset that is explained by a given number of principal components or factors

```python
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd

# Assuming 'test_final' is your DataFrame containing the dataset

# Create a PCA object with the desired number of components
pca = PCA(n_components=100, random_state=10)

# Fit the PCA model to your data and transform it
x_pca = pca.fit_transform(test_final.iloc[:, 0:-1])  # Excluding the target column

# Access explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Calculate the cumulative variance explained
cumulative_variance_explained = np.cumsum(explained_variance_ratio)

# Print cumulative variance explained
print("Cumulative Variance Explained:")
print(cumulative_variance_explained)
```

# Visualize Variance Ratio and Cumulative Variance Ratio explained by each principal component

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Fit PCA
#pca = PCA()
pca = PCA(n_components=100)
pca.fit_transform(test_final.iloc[:, 0:-1])
pca_variance = pca.explained_variance_ratio_

# Calculate cumulative variance
cumulative_variance = np.cumsum(pca_variance)

# Plot explained variance and cumulative explained variance
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.bar(range(len(pca_variance)), pca_variance, alpha=0.5, align='center', label='individual variance')
plt.xlabel('Principal components')
plt.ylabel('Variance ratio')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Ratio')
plt.title('Cumulative Variance Explained')
plt.grid()
plt.show()
```

## K-Nearest Neighbors (KNN) classification with PCA

This will split our data X and target y into training and testing sets. X_train and y_train will contain 60% of the data, and X_test and y_test will contain the remaining 40%. The random_state parameter is used to ensure reproducibility. We can change the random_state value to get different splits.

```python
# Import the necessary modules for ML model
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report

# Setup the pipeline steps:
steps = [('knn', KNeighborsClassifier())]

# Create the pipeline: pipeline
pipeline = Pipeline(steps)

# have paased less range value of hyperparamter since we are using free tier version of google colab.
k_range = list(range(1, 9))
parameters = dict(knn__n_neighbors=k_range)

#input
X = result_df.iloc[:,0:-1]

#target
y=result_df.iloc[:,-1]

# Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4,random_state=42)

#increasing cv score takes lot of time in gooogle colab, so we kept it just 2.
cv = GridSearchCV(pipeline,parameters,cv=2)

cv.fit(X_train,y_train)

# Predict the labels of the test set: y_pred
y_pred = cv.predict(X_test)

Knn_Accuracy = cv.score(X_test, y_test)

print("Training Set - X_train:")
print(X_train)

print("\nTesting Set - X_test:")
print(X_test)

print("\nTraining Set - y_train:")
print(y_train)

print("\nTesting Set - y_test:")
print(y_test)
```

**Accuracy Score : K-Nearest Neighbors (KNN) classification with PCA**

```python
# Compute and print metrics
print("Accuracy: {}".format(Knn_Accuracy))
print(classification_report(y_test, y_pred))
print("Tuned Model Parameters: {}".format(cv.best_params_))
```

Principal Component Analysis (PCA) is a dimensionality reduction technique commonly used in data analysis and machine learning. ## Ensemble

Ensemble learning is a machine learning technique that involves combining the predictions of multiple models to produce a more robust and accurate prediction than any individual model. The idea is that by combining diverse models, the strengths of one model can compensate for the weaknesses of another, leading to better overall performance. There are several types of ensemble methods, including bagging, boosting, and stacking.
